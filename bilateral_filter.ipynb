```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt

img = cv2.imread("D:\\picfolder\\fung.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# kernal using (1/25)*[ matrix]5*5
kernal = np.ones((5,5), np.float32)/25
bif= cv2.bilateralFilter(img, 30, 75, 75)

cv2.imshow("bilateral", bif)

#plt.imshow(img)
plt.imshow(bif)
plt.subplot(1,3,1),plt.imshow(img),plt.title("orginal image")
plt.subplot(1,3,2),plt.imshow(bif),plt.title("bilateral")
plt.savefig("bif.png")
```
